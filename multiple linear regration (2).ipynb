```python
cars = pd.read_csv("cars.csv")
cars.head()
```

Assumption in Multilinear Regression

1.Linearity:The relationship between the predictors(X) and the response(Y) is linear.
2.independence:Observations are independent of each other.
3.Homoscedasticity:The residuals(Y-Y_hat)exhibit contant variance at all lelels of the predictor.

```python
model1 = smf.ols('MPG~WT+VOL+SP+HP',data=cars).fit()
model1.summary()
```

## observations from model summary ##

```python
The R-squared and adjusted R-suared values are good and about 75% ao variability in Y is explained by X columns
The probability value with respect to F-statistic is close to zero indicating that all or someof X columns are significant
The p-values for VOL and WT are higher than 5% indicating some interaction issue among themselves whih need to be further explored
```
